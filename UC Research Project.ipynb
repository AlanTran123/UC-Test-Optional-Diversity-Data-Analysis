## Abstract

- The purpose of this project was to evaluate whether the University of California’s shift to test-optional admissions, implemented in Fall 2021, was associated with measurable changes in the racial diversity of UC applicants. Prior research has shown that standardized tests like the SAT are strongly correlated with socioeconomic and racial disparities, raising the question of whether removing test requirements might broaden the applicant pool. Motivated by this debate, we investigated whether diversity increased, decreased, or remained stable after the SAT/ACT requirement was eliminated.
- To carry out this analysis, we collected two publicly available datasets from the University of California: first-year applicant counts and transfer applicant counts, both broken down by race from 2016–2024. After cleaning and structuring the data, we visualized year-to-year changes using line plots and then calculated two standard diversity metrics: the Shannon Diversity Index and the Simpson Biodiversity Index (1 - D). These indices allowed us to quantify changes in racial representation before and after the test-optional policy. We defined meaningful change as ≥0.05 for Shannon and ≥0.02 for Simpson, based on accepted thresholds in diversity research.

- Our findings show that for freshman applicants, racial diversity remained remarkably stable across the policy change. Both diversity indices showed only extremely small decreases from 2020 to 2021, far below the thresholds needed to conclude meaningful change. For transfer applicants, the Shannon Index showed a notable decrease, while the Simpson Index showed a moderate decline. However, these shifts coincided with major external events—most significantly the COVID-19 pandemic’s impact on community college enrollment—which likely played a larger role than the test-optional policy itself.

- Overall, our analysis suggests that the transition to test-optional admissions was not associated with a meaningful change in the racial diversity of UC applicants, either for first-year or transfer populations. While some demographic shifts occurred, they were small, inconsistent across indices, or better explained by outside factors rather than the admissions policy itself. Therefore, based on the available data and predefined diversity thresholds, we conclude that the UC test-optional policy did not significantly alter applicant diversity during the years studied.

## Background and Prior Work

Our project revolves around the SAT, a standardized test that was supposedly used to objectively measure a student’s academic readiness. Reports from the Brookings Institution highlighted that these gaps were not a measure of aptitude, but rather a reflection of socioeconomic advantages. <a name="cite_ref-3"></a><sup>3</sup> This conversation ultimately led to a debate over standardized testing which led to a controversial “test-optional” policy for CA residents at UC Universities. Our project will analyze the UC’s public admissions data to see if the shift from “SAT required” to “SAT optional/blind” was associated with the change in the number and proportion, and the racial diversity of first year applicants.<br><br>
	To analyze this shift, we will be using both the Shannon Diversity Index and Simpson’s Diversity Index to see if there is a meaningful change in diversity in student applicant populations. For the Shannon Diversity Index, we will define this as a change of ΔH’ ≥ 0.05, and for the Simpson Diversity Index, we will define this as a change of Δ(1-D) ≥ 0.02. These thresholds are consistent with conventional interpretations in diversity measurement in other studies, and represent shifts large enough to suggest real change in population instead of random fluctuations.<a name="cite_ref-4"></a><sup>4</sup> Together, they should be able to confirm each other’s findings, and provide a basis for analysis.

### Exploratory Data Analysis - The Simpson and Shannon diversity Index

To evaluate whether racial diversity changed after UC adopted a test-optional admissions policy, we calculated two common diversity measures, the Shannon Index and the Simpson Index, using the racial applicant counts from 2016 to 2024. <br><br>

###### <b>The Simspon Index</b>
The Simpson's Biodiversity Index will measure the probability that two randomly chosen applicants belong to different racial groups. Higher values mean more diversity.

```python
def calculate_simpson(df):
    """
    Calculates Simpson Diversity Index (1-D) for each year (column).
    Formula: 1 - D = 1 - sum(p^2)
    """
    years = df.columns
    simpson_list = []
    
    for year in years:
        counts = df[year].values
        total = np.sum(counts)
        proportions = counts / total
        
        simpson = 1 - np.sum(proportions**2)
        simpson_list.append(simpson)
        
    return years, simpson_list
```

###### <b>The Shannon Index</b>
The Shannon Index measures how evenly represented/distributed different racial groups are. <br>

```python
def calculate_shannon(df):
    """
    Calculates Shannon Diversity Index (H') for each year (column).
    Formula: H = -sum(p * ln(p))
    """
    years = df.columns
    shannon_list = []
    
    for year in years:
        counts = df[year].values
        total = np.sum(counts)
        proportions = counts / total
        
        proportions_nonzero = proportions[proportions > 0]
        shannon = -np.sum(proportions_nonzero * np.log(proportions_nonzero))
        shannon_list.append(shannon)
        
    return years, shannon_list
```

```python
#Simpson Calculations
years_t, simpson_t = calculate_simpson(transfer_processed)
years_f, simpson_f = calculate_simpson(freshmen_processed)

#Shannon Calculations
years_t, shannon_t = calculate_shannon(transfer_processed)
years_f, shannon_f = calculate_shannon(freshmen_processed)
```

### First Analysis You Did - Comparing the Simpson Index between first-year and transfer applicants 

```python
# Simpson Plot
plt.figure(figsize=(12, 10))
plt.subplot(2, 1, 2)
plt.plot(years_t, simpson_t, marker='o', label='Transfer', linewidth=2, color='orange')
plt.plot(years_f, simpson_f, marker='s', label='Freshmen', linewidth=2, color='blue')
plt.axvline(x='2021', color='red', linestyle='--', label='Test-Optional Policy Start (2021)')
plt.axvline(x='2023', color='purple', linestyle='--', label='Test-Blind Policy Start (2023)')
plt.title("Simpson Diversity Index (1 - D)")
plt.ylabel('Simpson Index')
plt.xlabel('Year')
plt.legend()
plt.grid(True, alpha=0.3)
plt.show()
```
